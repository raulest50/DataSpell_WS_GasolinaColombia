```python
from sodapy import Socrata
import pandas as pd

# Conexión a la API de Socrata
client = Socrata("www.datos.gov.co", None)

P1 = "GASOLINA MOTOR CORRIENTE"
P2 = "BIODIESEL CON MEZCLA"
P3 = "GASOLINA MOTOR EXTRA"

C1 = "COMERCIALIZADOR INDUSTRIAL"
C2 = "ESTACION DE SERVICIO AUTOMOTRIZ"
C3 = "ESTACION DE SERVICIO FLUVIAL"

# ID del dataset: fbht-2fzd
dataset = "339g-zjac"
query = (
        f"SELECT SUM(volumen_despachado) as volumen_total, anio_despacho, mes_despacho, producto, departamento, municipio "
        f"WHERE subtipo_comprador IN ('{C2}') "
        f"AND producto IN ('{P2}') "
        f"GROUP BY anio_despacho, mes_despacho, producto, departamento, municipio "
        f"ORDER BY anio_despacho ASC "
        f"LIMIT 250000"
    )
consumo = client.get(dataset, query=query)
df_consumo = pd.DataFrame.from_records(consumo)
print(len(df_consumo))
```

```python
# Buscar registros de Medellín en Antioquia
medellin_antioquia = df_consumo[(df_consumo['municipio'] == 'MEDELLIN') & (df_consumo['departamento'] == 'ANTIOQUIA')]

# Verificar si existen registros
if len(medellin_antioquia) > 0:
    print(f"Se encontraron {len(medellin_antioquia)} registros para MEDELLIN, ANTIOQUIA")

    # Mostrar los primeros registros
    print("\nPrimeros registros encontrados:")
    print(medellin_antioquia.head())

    # Resumen por año y mes
    print("\nResumen de volumen total por año y mes:")
    resumen = medellin_antioquia.groupby(['anio_despacho', 'mes_despacho'])['volumen_total'].sum().reset_index()
    print(resumen)
else:
    print("No se encontraron registros para MEDELLIN, ANTIOQUIA en el dataset")

    # Verificar valores únicos en municipio y departamento para diagnosticar
    print("\nValores únicos de municipio en el dataset:")
    print(df_consumo['municipio'].unique())

    print("\nValores únicos de departamento en el dataset:")
    print(df_consumo['departamento'].unique())
```

```python
import numpy as np

# Función para crear una tabla con el consumo de diesel y la variación interanual
def crear_tabla_consumo_diesel(dataframe, nombre_tabla):
    # Filtrar solo para BIODIESEL CON MEZCLA
    diesel_data = dataframe[dataframe['producto'] == 'BIODIESEL CON MEZCLA'].copy()

    # Convertir la columna volumen_total a float antes de agrupar
    # Usamos coerce para que los valores problemáticos se conviertan en NaN
    diesel_data['volumen_total'] = pd.to_numeric(diesel_data['volumen_total'], errors='coerce')

    # Eliminar filas con NaN si es necesario
    diesel_data = diesel_data.dropna(subset=['volumen_total'])

    # Agrupar por año y calcular la suma del volumen
    consumo_anual = diesel_data.groupby('anio_despacho')['volumen_total'].sum().reset_index()

    # Convertir a millones de galones
    consumo_anual['volumen_millones'] = consumo_anual['volumen_total'] / 1000000

    # Calcular la variación interanual en porcentaje
    consumo_anual['variacion_porcentaje'] = consumo_anual['volumen_millones'].pct_change() * 100

    # Formato para presentación
    consumo_anual['volumen_formateado'] = consumo_anual['volumen_millones'].apply(lambda x: f"{x:.2f}")
    consumo_anual['variacion_porcentaje_formateada'] = consumo_anual['variacion_porcentaje'].apply(
        lambda x: f"{x:+.2f}%" if not np.isnan(x) else "N/A"
    )

    # Renombrar columnas para mejor presentación
    tabla_final = consumo_anual[['anio_despacho', 'volumen_formateado', 'variacion_porcentaje_formateada']]
    tabla_final.columns = ['Año', 'Volumen (millones gal)', 'Variación Interanual (%)']

    print(f"\n{'-'*20} {nombre_tabla} {'-'*20}")
    print(tabla_final.to_string(index=False))

    return consumo_anual

# Verificar que existan datos de diesel en cada dataframe
def verificar_datos_diesel(dataframe, nombre):
    tiene_diesel = 'BIODIESEL CON MEZCLA' in dataframe['producto'].unique()
    if not tiene_diesel:
        print(f"\n{'-'*20} {nombre} {'-'*20}")
        print(f"No hay datos de BIODIESEL CON MEZCLA en {nombre}")
        print(f"Productos disponibles: {dataframe['producto'].unique()}")
    return tiene_diesel

# Primero, verificar y limpiar la columna volumen_total en los dataframes
print("\nVerificando y limpiando datos...")

# Limpiar df
df['volumen_total'] = pd.to_numeric(df['volumen_total'], errors='coerce')
print(f"Datos nivel nacional: {len(df)} registros totales, {df['volumen_total'].isna().sum()} valores nulos")

# Crear los dataframes filtrados (en caso de que no existan)
if 'df_antioquia' not in locals():
    df_antioquia = df[df['departamento'] == 'ANTIOQUIA'].copy()

if 'df_medellin' not in locals():
    df_medellin = df[df['municipio'] == 'MEDELLIN'].copy()

# Limpiar df_antioquia
df_antioquia['volumen_total'] = pd.to_numeric(df_antioquia['volumen_total'], errors='coerce')
print(f"Datos Antioquia: {len(df_antioquia)} registros totales, {df_antioquia['volumen_total'].isna().sum()} valores nulos")

# Limpiar df_medellin
df_medellin['volumen_total'] = pd.to_numeric(df_medellin['volumen_total'], errors='coerce')
print(f"Datos Medellín: {len(df_medellin)} registros totales, {df_medellin['volumen_total'].isna().sum()} valores nulos")

# Eliminar filas con valores nulos para evitar problemas en los cálculos
df = df.dropna(subset=['volumen_total'])
df_antioquia = df_antioquia.dropna(subset=['volumen_total'])
df_medellin = df_medellin.dropna(subset=['volumen_total'])

# Crear las tablas para cada dataframe
print("\nTABLAS DE CONSUMO DE BIODIESEL Y VARIACIÓN INTERANUAL")

# Para el nivel nacional (df)
if verificar_datos_diesel(df, "NIVEL NACIONAL"):
    datos_nacional = crear_tabla_consumo_diesel(df, "NIVEL NACIONAL")

# Para Antioquia (df_antioquia)
if len(df_antioquia) > 0 and verificar_datos_diesel(df_antioquia, "ANTIOQUIA"):
    datos_antioquia = crear_tabla_consumo_diesel(df_antioquia, "ANTIOQUIA")

# Para Medellín (df_medellin)
if len(df_medellin) > 0 and verificar_datos_diesel(df_medellin, "MEDELLÍN"):
    datos_medellin = crear_tabla_consumo_diesel(df_medellin, "MEDELLÍN")

# Resumen comparativo (solo si hay datos para todos)
try:
    if ('datos_nacional' in locals() and
        'datos_antioquia' in locals() and
        'datos_medellin' in locals() and
        not datos_nacional.empty and
        not datos_antioquia.empty and
        not datos_medellin.empty):

        print("\n\nRESUMEN COMPARATIVO DE VARIACIÓN INTERANUAL (%)")

        # Unir los datos para comparación
        años_comunes = sorted(set(datos_nacional['anio_despacho']) &
                             set(datos_antioquia['anio_despacho']) &
                             set(datos_medellin['anio_despacho']))

        comparativo = pd.DataFrame({'Año': años_comunes})

        # Función para obtener la variación porcentual para un año dado
        def obtener_variacion(df_datos, año):
            fila = df_datos[df_datos['anio_despacho'] == año]
            if len(fila) > 0:
                return fila['variacion_porcentaje'].values[0]
            return np.nan

        # Agregar los datos de variación para cada región
        for año in años_comunes:
            comparativo.loc[comparativo['Año'] == año, 'Nacional'] = obtener_variacion(datos_nacional, año)
            comparativo.loc[comparativo['Año'] == año, 'Antioquia'] = obtener_variacion(datos_antioquia, año)
            comparativo.loc[comparativo['Año'] == año, 'Medellín'] = obtener_variacion(datos_medellin, año)

        # Formatear para presentación
        for col in ['Nacional', 'Antioquia', 'Medellín']:
            comparativo[col] = comparativo[col].apply(
                lambda x: f"{x:+.2f}%" if not np.isnan(x) else "N/A"
            )

        print(comparativo.to_string(index=False))

        # Generar también un resumen de volúmenes (no solo variaciones)
        print("\n\nRESUMEN COMPARATIVO DE VOLUMEN (MILLONES DE GALONES)")

        volumen_comparativo = pd.DataFrame({'Año': años_comunes})

        # Función para obtener el volumen para un año dado
        def obtener_volumen(df_datos, año):
            fila = df_datos[df_datos['anio_despacho'] == año]
            if len(fila) > 0:
                return fila['volumen_millones'].values[0]
            return np.nan

        # Agregar los datos de volumen para cada región
        for año in años_comunes:
            volumen_comparativo.loc[volumen_comparativo['Año'] == año, 'Nacional'] = obtener_volumen(datos_nacional, año)
            volumen_comparativo.loc[volumen_comparativo['Año'] == año, 'Antioquia'] = obtener_volumen(datos_antioquia, año)
            volumen_comparativo.loc[volumen_comparativo['Año'] == año, 'Medellín'] = obtener_volumen(datos_medellin, año)

        # Formatear para presentación
        for col in ['Nacional', 'Antioquia', 'Medellín']:
            volumen_comparativo[col] = volumen_comparativo[col].apply(
                lambda x: f"{x:.2f}" if not np.isnan(x) else "N/A"
            )

        print(volumen_comparativo.to_string(index=False))

except Exception as e:
    print(f"\nNo se pudo generar el resumen comparativo: {str(e)}")
```
